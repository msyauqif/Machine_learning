```python
#IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('botak.csv')
```

```python
df
```

## Diagram Batang: Rata-rata Gaji Berdasarkan Pendidikan

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x="pendidikan", y="gaji", data=df, palette="coolwarm", hue=None, ax=ax)
ax.set_title("Rata-rata Gaji Berdasarkan Pendidikan", fontsize=14)
ax.set_xlabel("Pendidikan", fontsize=12)
ax.set_ylabel("Rata-rata Gaji", fontsize=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, fontsize=12)
fig.text(0.5, 0.5, "Muhammad Syauqi Firdaus", fontsize=30, color='gray', alpha=0.2, ha='center', va='center', rotation=30)
plt.show()
```

## Diagram Garis: Tren Rata-rata Tinggi Badan Berdasarkan Usia

```python
umur_tinggi = df.groupby("umur")["tinggi"].mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 5))
sns.lineplot(x="umur", y="tinggi", data=umur_tinggi, marker="o", color="green", ax=ax)
ax.set_title("Tren Rata-rata Tinggi Badan Berdasarkan Usia", fontsize=14)
ax.set_xlabel("Umur", fontsize=12)
ax.set_ylabel("Tinggi Badan (cm)", fontsize=12)
fig.text(0.5, 0.5, "Muhammad Syauqi Firdaus", fontsize=30, color='gray', alpha=0.2, ha='center', va='center', rotation=30)
plt.show()
```

## Diagram Lingkaran: Proporsi Responden Berdasarkan Jenis Kelamin

```python
fig, ax = plt.subplots(figsize=(6, 6))
df["jenis_kelamin"].value_counts().plot.pie(autopct='%1.1f%%', colors=["red", "pink"], ax=ax)
ax.set_title("Proporsi Responden Berdasarkan Jenis Kelamin", fontsize=14)
fig.text(0.5, 0.5, "Muhammad Syauqi Firdaus", fontsize=30, color='gray', alpha=0.2, ha='center', va='center', rotation=30)
ax.set_ylabel("")
plt.show()
```
